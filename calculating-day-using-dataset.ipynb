```python
import pandas as pd
import json

# Load your CSV data into a DataFrame
data = pd.read_csv('/kaggle/input/day-1-01/day1_03.csv')

# Define a list of columns that are present in your dataset
valid_columns = [
    'open', 'high', 'low', 'close', 'volume', 'day', 'month', 'year',
    'Volatility_High', 'Volatility_Low', 'Trend_Duration', 'RSI', 'MACD', 'Signal_Line',
    'ATR', 'ADX', '%K', '%D', 'ROC', 'ULTOSC', 'CCI', 'BOP', 'TR', 'TRavg', 'CMO',
    'Price_Up_3Days', 'Price_Up_7Days', 'Price_Up_15Days', 'Price_Up_1Month',
    'Price_Up_3Days_Direction', 'Price_Up_7Days_Direction', 'Price_Up_15Days_Direction', 'Price_Up_1Month_Direction', 'd1'
]

# Create a list to store JSON objects
json_objects = []

# Define a function to handle blank values
def handle_value(row, column):
    value = row.get(column, "N/A")
    return value

# Iterate over your dataset and create JSON objects
for index, row in data.iterrows():
    # Create the prompt text based on the valid columns
    prompt = f"On {row['month']}/{row['day']}/{row['year']}, the EUR/USD forex market opened at a rate of {handle_value(row, 'open')} and traded within a range of {handle_value(row, 'low')} (low) to {handle_value(row, 'high')} (high). The trading volume was {handle_value(row, 'volume')}. Technical Indicators: RSI - {handle_value(row, 'RSI')}, MACD - {handle_value(row, 'MACD')}, ATR - {handle_value(row, 'ATR')}, ADX - {handle_value(row, 'ADX')}, %K - {handle_value(row, '%K')}, %D - {handle_value(row, '%D')}, CCI - {handle_value(row, 'CCI')}, BOP - {handle_value(row, 'BOP')}, TR - {handle_value(row, 'TR')}"

    # Create the completion text based on the valid columns
    completion = f"The EUR/USD price for the next day has gone {handle_value(row, 'd1')}. Over the last 3 days, the EUR/USD price has gone {handle_value(row, 'Price_Up_3Days')} ({handle_value(row, 'Price_Up_3Days_Direction')}). Over the last 7 days, the EUR/USD price has gone {handle_value(row, 'Price_Up_7Days')} ({handle_value(row, 'Price_Up_7Days_Direction')}). Over the last 15 days, the EUR/USD price has gone {handle_value(row, 'Price_Up_15Days')} ({handle_value(row, 'Price_Up_15Days_Direction')}). Over the last month, the EUR/USD price has gone {handle_value(row, 'Price_Up_1Month')} ({handle_value(row, 'Price_Up_1Month_Direction')}) based on the day time frame."

    json_object = {
        "prompt": prompt,
        "completion": completion
    }
    json_objects.append(json_object)

# Write the JSON objects to a JSONL file
with open('/kaggle/working/formatted_data.jsonl', 'w') as file:
    for json_object in json_objects:
        json.dump(json_object, file)
        file.write('\n')
```

```python
import pandas as pd
import json

# Load your CSV data into a DataFrame
data = pd.read_csv('/kaggle/input/day-1-01/day1_03.csv')

# Create a list to store JSON objects
json_objects = []

# Iterate over your dataset and create JSON objects
for index, row in data.iterrows():
    prompt_text = f"On {row['month']}/{row['day']}/{row['year']}, the EUR/USD forex market opened at a rate of {row['open']} and traded within a range of {row['low']} (low) to {row['high']} (high). The trading volume was {row['volume']}. Technical Indicators:"

    completion_text = f"The EUR/USD price for the next day has gone {row['d1']}. Over the last 3 days, the EUR/USD price has gone {row['Price_Up_3Days']} ({row['Price_Up_3Days_Direction']}). Over the last 7 days, the EUR/USD price has gone {row['Price_Up_7Days']} ({row['Price_Up_7Days_Direction']}). Over the last 15 days, the EUR/USD price has gone {row['Price_Up_15Days']} ({row['Price_Up_15Days_Direction']}). Over the last month, the EUR/USD price has gone {row['Price_Up_1Month']} ({row['Price_Up_1Month_Direction']}) based on the day time frame."

    # Exclude columns with no values
    for col in data.columns:
        if pd.notna(row[col]):
            prompt_text += f" {col} - {row[col]},"
            completion_text += f" {col} - {row[col]},"

    # Remove the trailing comma and spaces
    prompt_text = prompt_text.rstrip(', ')
    completion_text = completion_text.rstrip(', ')

    json_object = {
        "prompt": prompt_text,
        "completion": completion_text
    }
    json_objects.append(json_object)

# Write the JSON objects to a JSONL file
with open('/kaggle/working/day1_04.jsonl', 'w') as file:
    for json_object in json_objects:
        json.dump(json_object, file)
        file.write('\n')
```
