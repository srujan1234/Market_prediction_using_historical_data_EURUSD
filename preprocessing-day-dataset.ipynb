```python
import pandas as pd

# Load your dataset from the CSV file with tab as the delimiter
df = pd.read_csv('/kaggle/input/day-dataset/D1.csv', delimiter='\t')

# Calculate the 3-Day Price_Up
df['Price_Up_3Days'] = df['close'].sub(df['close'].shift(3))

# Calculate the 7-Day Price_Up
df['Price_Up_7Days'] = df['close'].sub(df['close'].shift(7))

# Calculate the 15-Day Price_Up
df['Price_Up_15Days'] = df['close'].sub(df['close'].shift(15))

# Calculate the 1-Month Price_Up (30 days)
df['Price_Up_1Month'] = df['close'].sub(df['close'].shift(30))

# Save the updated DataFrame to a new CSV file
df.to_csv('/kaggle/working/D1_with_Price_Up.csv', index=False)
```

```python
import pandas as pd
import numpy as np
import json

# Load your dataset into a Pandas DataFrame
df = pd.read_csv('/kaggle/input/cleaned-ready-d1/D1.csv', delimiter='\t')

# Define a function to create prompt-completion pairs for each row
def generate_prompt_completion(row):
    # Define the column names based on your dataset
    open_price = row['open']
    high_price = row['high']
    low_price = row['low']
    close_price = row['close']
    volume = row['volume']
    day = row['day']
    month = row['month']
    year = row['year']
    trend_duration = row['Trend_Duration']
    rsi = row['RSI']
    macd = row['MACD']
    signal_line = row['Signal_Line']
    atr = row['ATR']
    adx = row['ADX']
    k_percent = row['%K']
    d_percent = row['%D']
    mfi = row['MFI']
    roc = row['ROC']
    ultosc = row['ULTOSC']
    cci = row['CCI']
    bop = row['BOP']
    tr = row['TR']
    tr_avg = row['TRavg']
    cmo = row['CMO']
    price_up_3days = row['Price_Up_3Days']
    price_up_7days = row['Price_Up_7Days']
    price_up_15days = row['Price_Up_15Days']
    price_up_1month = row['Price_Up_1Month']
    price_up_3days_direction = row['Price_Up_3Days_Direction']
    price_up_7days_direction = row['Price_Up_7Days_Direction']
    price_up_15days_direction = row['Price_Up_15Days_Direction']
    price_up_1month_direction = row['Price_Up_1Month_Direction']

    # Handle missing values
    open_price = f"{open_price:.5f}" if not np.isnan(open_price) else "N/A"
    high_price = f"{high_price:.5f}" if not np.isnan(high_price) else "N/A"
    low_price = f"{low_price:.5f}" if not np.isnan(low_price) else "N/A"
    close_price = f"{close_price:.5f}" if not np.isnan(close_price) else "N/A"
    volume = f"{volume:.2f}" if not np.isnan(volume) else "N/A"

    # Create the prompt text
    prompt_text = f"On {month}/{day}/{year}, the EUR/USD forex market opened at a rate of {open_price} and traded within a range of {low_price} (low) to {high_price} (high). The trading volume was {volume}."

    # Create the completion text (you can customize this based on your template)
    completion_text = f"The next day's EUR/USD price has gone {price_up_1month} ({price_up_1month_direction}). Over the last 3 days, the EUR/USD price has gone {price_up_3days} ({price_up_3days_direction}). Over the last 7 days, the EUR/USD price has gone {price_up_7days} ({price_up_7days_direction}). Over the last 15 days, the EUR/USD price has gone {price_up_15days} ({price_up_15days_direction}) based on the day time frame."

    return prompt_text, completion_text

# Create a list to store the prompt-completion pairs
prompt_completion_pairs = []

# Process each row in the DataFrame
for index, row in df.iterrows():
    prompt_text, completion_text = generate_prompt_completion(row)
    pair = {
        "prompt": prompt_text,
        "completion": completion_text
    }
    prompt_completion_pairs.append(pair)

# Save the prompt-completion pairs to a JSONL file
with open('prompt_completion_pairs.jsonl', 'w', encoding='utf-8') as file:
    for pair in prompt_completion_pairs:
        json.dump(pair, file)
        file.write('\n')

print("Prompt-completion pairs saved to 'prompt_completion_pairs.jsonl'.")
```
